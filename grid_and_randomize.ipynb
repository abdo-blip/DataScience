```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
breast_cancer_dataset = datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
breast_data = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
breast_data.head()
```

```python
breast_data['target'] = breast_cancer_dataset.target
```

```python
breast_data.head()
```

```python
breast_data['target'].value_counts()
```

```python
breast_data.isnull().sum()
```

```python
x = breast_data.drop(columns = ['target'], axis = 1)
y = breast_data['target']
```

```python
X = np.array(x)
Y = np.array(y)
```

```python
model = SVC()
```

```python
# hyperparameters

parameters = {
    
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'C' : [1, 5, 10, 20]
    
    
}
```

```python
classifier = GridSearchCV(model, parameters, cv = 5)
```

```python
classifier.fit(x, y)
```

```python
print(classifier.cv_results_)
```

```python
best_parameters = classifier.best_params_
best_estimators = classifier.best_estimator_
print(best_parameters)
print(best_estimators)
```

```python
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
result['param_kernel' ].value_counts()
```

```python
grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
print(grid_search_result)
```

```python
randomize = RandomizedSearchCV(model, parameters, cv = 5)
```

```python
randomize.fit(x, y)
```

```python
print(randomize.cv_results_)
```

```python
print(randomize.best_params_)
print(randomize.best_estimator_)
```

```python
print(randomize.best_score_)
```

```python
result = pd.DataFrame(randomize.cv_results_)
```

```python
result.head()
```
